```python
# Project A_info1998 (clean Data)
import pandas as pd 
import numpy as np
# import dataset 
filepath = 'D:/Cornell/FALL2017/info1998/Project1/train.csv'
house = pd.read_csv(filepath)
# Baisc Clean Data in ProjectA
med = house['LotFrontage'].describe()['50%']
house['LotFrontage'] = house['LotFrontage'].fillna(value = med)
# Alley
del house['Alley']
# MasVnrArea
med = house['MasVnrArea'].describe()['50%']
house['MasVnrArea'] = house['MasVnrArea'].fillna(value = med)
# FireplaceQu
del house['FireplaceQu']
# GarageType
house['GarageType'] = house['GarageType'].fillna(value = -200)
# GarageYrBlt
house['GarageYrBlt'] = house['GarageYrBlt'].fillna(value = med)
# PoolQC
del house['PoolQC']
# Fence
del house['Fence']
# MiscFeature
del house['MiscFeature']
# fill other all na value to -200 
house_filled = house.fillna(value=-200.0)
# check all missing data 
sum(house_filled.isnull().any())
```

## Logistics Regression

```python
#exclude 5 in OverallQual column
house = house[house.OverallQual !=5]
```

```python
#do correlation test to choose predictors
house.corr()
```

```python
# Logistics regression
# set predictors
X = pd.DataFrame(house[['YearBuilt','YearRemodAdd','GarageCars','SalePrice']])
#tranform predictors to categorical integer values
enc = LabelEncoder()
for i in X.columns:
    X[i] = enc.fit_transform(X[i])
# set response variable Y=OverallQual  >= average as 1 and < average as 0 
Y  = (house['OverallQual'].values >= np.mean(house['OverallQual'])) + 0 
```

```python
# import necessary packages
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
#baseline calculation
#calculated as the percentage of the column's most frequent value
sum(Y == 1) > sum(Y == 0)
baseline = float(sum(Y == 1))/len(Y)
print'baseline: ', 100.0*baseline, '%'
```

```python
random_state=0
for i in range(5):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = random_state)
    
    logit = LogisticRegression()
    logit.fit(X_train, Y_train)
    
    # make predictions on test set
    pred_logit = logit.predict(X_test)
    accuracy=accuracy_score(y_true = Y_test, y_pred = pred_logit)
    print 'Split',i+1,'accuracy rate is:',accuracy,'. Compared to the baseline, it is',(accuracy-baseline)/baseline*100,'% more accurate.'
    
    
    random_state+=1
```

Paragraphï¼š At first I exclude the OverallQual value that's equals to 5 as it required in the project description. Then I did correlation check to pick my predictors as the variables most correlated to my response variable, including 'YearBuilt','YearRemodAdd','GarageCars' and 'SalePrice'. In the next step, I transformed them into categorical integer values for conducting logistic regression. After that, I calculated the baseline as the percentage of the columns' most frequent value, which is 51.55%. Next, I write a for loop to conduct logistic regression five different split of train/test dataset regarding to different random states. My result turns out that logistic regression on different data split are all more than 40% accurate than baseline.

## SVM Model

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
## labelize 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(house['HouseStyle'])
## unique levels for houseStyle
print('Unique classifications:\n',list(le.classes_))
y = le.transform(house['HouseStyle'])
```

```python
# baseline (it is equivalent to the percentage of the column's most frequent value.)
## count the frequency for each classifications 
print('Frequency table:\n',pd.DataFrame({'y':list(y)}).y.value_counts())
print()
baseline = round(sum(y == 2)/len(y),3)
print('Baseline:',baseline)
```

```python
# MSSubClass, 2ndFlrSF ,HalfBath
dataset = house
dataset['HouseStyle'] = y
dataset.corr()['HouseStyle'] > 0.35
## choose the variables with correlation greater than 0.35. 
X = dataset[['MSSubClass','2ndFlrSF','HalfBath']]
```

```python
for i in range(1,6):
    print('Iter',i,':\n')
    print('test_size split:', str(round(0.1 * i,2)))
    print('random_state:', i)
    classifier = svm.SVC(kernel = 'rbf',C = 1)
    ## splot the training and testing data 
    X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size = 0.1, random_state = i)
    ## fit the model to the training data 
    classifier.fit(X_train, Y_train)
    # predict the y_test 
    pred = classifier.predict(X_test)
    # accuray score
    score = classifier.score(X_test,Y_test)
    improve = (score - baseline)/baseline
    print('The accuracy score: ',score, '\n')
    print('The improvement based on Baseline: ',improve, '\n')
    print('\n\n')
```

The SVM model is applied to predict the variable of HouseStyle with the several parameters. Before Building the SVM model, firstly I transformed the predicted valuable of HouseStyle to 8 catergorical numerical levels 0 to 7. The variables of MSSubClass, 2ndFlrSF ,HalfBath are chosen due the correlation with the predicted variable which the correlations are greater than 0.35. And also, I calculated the baseline of HouseStyle as 0.49. All five SVM models I built are 75% more than accuracy of Baseline.

```python
# baseline (it is equivalent to the percentage of the column's most frequent value.)
# count the frequency for each classifications 
print('Frequency table:\n',pd.DataFrame({'Y':list(Y)}).Y.value_counts())
print()
baseline = float(sum(Y == 'CollgCr'))/len(Y)
print('Baseline: %f' % baseline)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X = np.array(house[['OverallQual', 'YearBuilt', 'LotArea']])
Y = house['Neighborhood']

random_state = 0
for i in range(5):

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = random_state)

    # build model and fit on train set
    dTree = DecisionTreeClassifier(max_leaf_nodes=15)
    dTree.fit(X_train, Y_train)

    random_state += 1
    
    dTree_pred = dTree.predict(X_test)
    dTree_pred
    
    accuracy = accuracy_score(y_true = Y_test, y_pred = dTree_pred)
    print('Split',i+1,'accuracy rate is:',accuracy,'. Compared to the baseline, it is',((accuracy-baseline)/baseline) * 100,'% more accurate.')
```

To build out the decision tree model, we have selected the following independent variables: OverallQual, YearBuilt, LotArea; all of these are critical to predict Neighborhood as an outcome. Our selection rationale is based on the relevancy of the predictor variables. For OverallQual, the level of rating helps to determine the type of neighborhood; YearBuilt is useful to categorize the time periods in which the neighborhoods were built; Lastly, the variable of LotArea might help to signal the type of neighborhood as an attribute.
